```python
#Using Canny edge detector
# and matplotlib to to display the image

import cv2
import matplotlib.pyplot as plt
```

```python
# imread() is a function used to read the image and the load a a grayscale background
image = cv2.imread('path/to/image', cv2.COLOR_BGR2GRAY)

# passed parameters low_thresh refers to the minimum gradient required to consider a pixel as an edge
# high_thresh - maximum gradient to consider pixel an edge
# 3 is aperture size (sorbel kernel)
# L2gradient
Low_thresh = 50
High_thresh = 150
edges = cv2.Canny(image, Low_thresh, High_thresh, 3, L2gradient=True)
plt.figure()
plt.title('Edges')
plt.imshow(edges, cmap='gray')
plt.show
```
